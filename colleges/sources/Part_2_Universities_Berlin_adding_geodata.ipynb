```python
import pandas as pd
import googlemaps
import time

# Load your CSV file into a DataFrame
df = pd.read_csv('universities_berlin_rating_final.csv')

# Initialize the Google Maps client with your API key
gmaps = googlemaps.Client(key='Insert Google Key Number here')

# Function to geocode an address using Google Maps Geocoding API
def geocode_address(address):
    try:
        geocode_result = gmaps.geocode(address)
        if geocode_result:
            location = geocode_result[0]['geometry']['location']
            postcode = next((addr['long_name'] for addr in geocode_result[0]['address_components'] if 'postal_code' in addr['types']), None)
            neighborhood = next((addr['long_name'] for addr in geocode_result[0]['address_components'] if 'neighborhood' in addr['types']), None)
            return {
                'latitude': location['lat'],
                'longitude': location['lng'],
                'postcode': postcode,
                'neighborhood': neighborhood
            }
    except Exception as e:
        print(f"Error geocoding {address}: {e}")
    return None

# Add new columns for geolocation data
df['latitude'] = None
df['longitude'] = None
df['postcode'] = None
df['neighborhood'] = None

# Iterate over the DataFrame and geocode each university
for index, row in df.iterrows():
    university_name = row['University Name']
    address = f"{university_name}, Berlin, Germany"
    result = geocode_address(address)
    if result:
        df.at[index, 'latitude'] = result.get('latitude')
        df.at[index, 'longitude'] = result.get('longitude')
        df.at[index, 'postcode'] = result.get('postcode')
        df.at[index, 'neighborhood'] = result.get('neighborhood')
    time.sleep(1)  # Rate limiting: wait for 1 second between requests

# Save the updated DataFrame to a new CSV file
df.to_csv('universities_berlin_rating_final_with_geolocation.csv', index=False)
```

```python
import pandas as pd
import googlemaps
import time

# Load your CSV file into a DataFrame
df = pd.read_csv('universities_berlin_rating_final_with_geolocation.csv')

# Initialize the Google Maps client
gmaps = googlemaps.Client(key='Google Key Number')

# Function to perform reverse geocoding and extract neighborhood information
def get_neighborhood(lat, lng):
    try:
        reverse_geocode_result = gmaps.reverse_geocode((lat, lng))
        for result in reverse_geocode_result:
            for component in result['address_components']:
                if 'sublocality_level_1' in component['types']:
                    return component['long_name']
        return None
    except Exception as e:
        print(f"Error in reverse geocoding: {e}")
        return None

# Apply the reverse geocoding function to each row
df['neighborhood'] = df.apply(lambda row: get_neighborhood(row['latitude'], row['longitude']), axis=1)

# Save the updated DataFrame to a new CSV file
df.to_csv('universities_Berlin_full_info.csv', index=False)

# Display the updated DataFrame
print("Final DataFrame with neighborhood information:")
print(df)
```

```python
import pandas as pd

# Load your CSV file into a DataFrame
df = pd.read_csv('universities_Berlin_full_info_clean.csv')

# Mapping of neighborhoods to AGS codes
ags_mapping = {
    'Mitte': '11000001',
    'Friedrichshain-Kreuzberg': '11000002',
    'Pankow': '11000003',
    'Charlottenburg-Wilmersdorf': '11000004',
    'Spandau': '11000005',
    'Steglitz-Zehlendorf': '11000006',
    'Tempelhof-Schöneberg': '11000007',
    'Neukölln': '11000008',
    'Treptow-Köpenick': '11000009',
    'Marzahn-Hellersdorf': '11000010',
    'Lichtenberg': '11000011',
    'Reinickendorf': '11000012'
}

# Function to get AGS code from neighborhood
def get_ags_code(neighborhood):
    return ags_mapping.get(neighborhood, None)

# Apply the function to get AGS code
df['AGS_Code'] = df['neighborhood'].apply(get_ags_code)

# Ensure AGS_Code is a string type
df['AGS_Code'] = df['AGS_Code'].astype(str)

# Transform postcode to string and remove trailing zeros and decimal points
if 'postcode' in df.columns:
    df['postcode'] = df['postcode'].astype(str).str.replace(r'\.0$', '', regex=True)

# Save the updated DataFrame to a new CSV file
df.to_csv('universities_Berlin_final.csv', index=False)

# Display the updated DataFrame
print("Final DataFrame with AGS code and cleaned postcode information:")
print(df)
```
