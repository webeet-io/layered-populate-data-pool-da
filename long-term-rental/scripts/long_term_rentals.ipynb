```python
# Install required packages if not already installed (geopy and requests were installed earlier)
# !pip install geopy requests

import pandas as pd
from geopy.geocoders import Nominatim
import time
import requests

# Use Nominatim API with rate limiting
geolocator = Nominatim(user_agent="immowelt_colab_geocoder", timeout=10) # Using a user agent and timeout

def batch_geocode_details(df, batch_size=100): # Reduced batch size for potentially better stability
    results = []

    for i in range(0, len(df), batch_size):
        batch = df.iloc[i:i+batch_size]
        batch_results = []

        for idx, row in batch.iterrows():
            latitude = row['latitude']
            longitude = row['longitude']

            try:
                location = geolocator.reverse((latitude, longitude), exactly_one=True)
                if location and location.raw and 'address' in location.raw:
                    address = location.raw['address']
                    # Nominatim can return different address components, try common ones
                    neighbourhood = address.get('neighbourhood') or address.get('suburb') or address.get('city_district')
                    district = address.get('city_district') or address.get('suburb') or address.get('neighbourhood') or address.get('county')
                    batch_results.append({'neighbourhood': neighbourhood, 'district': district})
                else:
                    print(f"Could not geocode ({latitude}, {longitude}): No address found.")
                    batch_results.append(None)
            except (GeocoderTimedOut, GeocoderUnavailable) as e:
                 print(f"Geocoding service error for ({latitude}, {longitude}): {e}")
                 batch_results.append(None)
            except Exception as e:
                 print(f"An unexpected error occurred during geocoding for ({latitude}, {longitude}): {e}")
                 batch_results.append(None)

            # Add a small delay between requests within a batch
            time.sleep(1.5) # Increased delay slightly

        results.extend(batch_results)
        print(f"Processed batch {i//batch_size + 1}/{len(df)//batch_size + 1}")
        # Add a longer delay between batches
        time.sleep(5)

    return results

# Apply the batch geocoding function to the dataframe
geocoded_results = batch_geocode_details(df)

# Create new columns for the geocoded neighborhood and district from the results
df['geocoded_neighbourhood_nominatim'] = [x['neighbourhood'] if x else None for x in geocoded_results]
df['geocoded_district_nominatim'] = [x['district'] if x else None for x in geocoded_results]

display(df.head())
display(df.info())
```

```python
# Remove the 'district' column from the DataFrame
df = df.drop('district', axis=1)

display(df.head())
display(df.info())
```

```python
df = df.rename(columns={'geocoded_neighbourhood_nominatim': 'neighborhood', 'geocoded_district_nominatim': 'district'})
display(df.info())
```

```python
# Export the cleaned and transformed DataFrame to a CSV file
df.to_csv('long_term_rentals.csv', index=False)
```
